```python
import datetime
holiday = [
    "2024-01-01",
    "2024-02-09",
    "2024-02-10",
    "2024-02-11",
    "2024-02-12",
    "2024-03-01",
    "2024-05-05",
    "2024-05-06",
    "2024-06-06",
    "2024-08-15",
    "2024-09-16",
    "2024-09-17",
    "2024-09-18",
    "2024-10-03",
    "2024-10-09",
    "2024-12-25",
    "2023-01-01",
    "2023-01-21",
    "2023-01-22",
    "2023-01-23",
    "2023-01-24",
    "2023-03-01",
    "2023-05-05",
    "2023-05-27",
    "2023-05-29",
    "2023-06-06",
    "2023-08-15",
    "2023-09-28",
    "2023-09-29",
    "2023-09-30",
    "2023-10-02",
    "2023-10-03",
    "2023-10-09",
    "2023-12-25",
]

def get_weekday_name(date_str, date_format='%Y-%m-%d'):
    """
    날짜 문자열(date_str)을 입력받아 요일(한국어, 예: 월요일, 화요일, ...)을 반환합니다.
    date_format은 기본값 '%Y-%m-%d'를 사용합니다.
    """
    if date_str in holiday:
        return "일요일"
    days_kr = ['평일', '평일', '평일', '평일', '평일', '토요일', '일요일']
    date_obj = datetime.datetime.strptime(date_str, date_format)
    weekday = date_obj.weekday()  # 0=월요일, 6=일요일
    return days_kr[weekday]
```

```python
def get_previous_station_order(hosun: int, order_now: int,direction: str): 
    if hosun == 2:
        if direction == "오름차순":
            if order_now == 1:
                return 43
            else:
                return order_now - 1
        elif direction == "내림차순":
            if order_now == 43: 
                return 1
            else:
                return order_now + 1
    elif hosun == 3:
        if direction == "오름차순":
            if order_now == 1: 
                return None
            else:
                return order_now - 1
        elif direction == "내림차순":
            if order_now == 34:
                return None
            else:
                return order_now + 1
```

```python
station_2 = station[station["호선"] == 2]
station_3 = station[station["호선"] == 3]
def get_station_number(hosun: int, order: int): 
    if hosun == 2:
        return int(station_2[station_2["순서"] == order]["역번호"].values[0])
    elif hosun == 3:
        return int(station_3[station_3["순서"] == order]["역번호"].values[0])
```

```python
def process_boarding_data(boarding_data, station_data, congest_data):
    # Station 데이터의 "순서", "환승역" 병합
    boarding_new = boarding_data.copy()
    boarding_new = pd.merge(
        boarding_new,
        station_data[["역번호", "순서", "환승역"]],
        on="역번호",
        how="left"
    )
    # 데이터 타입 변환, NULL 값 처리
    boarding_new = boarding_new.dropna()
    boarding_new["환승역"] = boarding_new["환승역"].astype(int)
    boarding_new["순서"] = boarding_new["순서"].astype(int)

    # 요일 구분 컬럼 생성
    boarding_new["요일구분"] = boarding_new["날짜"].apply(get_weekday_name)

    # 혼잡도 컬럼 생성
    boarding_new["오름혼잡도"] = None
    boarding_new["내림혼잡도"] = None

    congest_lookup = congest_data.set_index(["요일구분", "시간대", "역번호", "상하구분"]).sort_index()

    def get_congestion_value(row, direction):
        key = (row["요일구분"], row["시간대"], row["역번호"], direction)
        try:
            return congest_lookup.loc[key, "혼잡도"]
        except KeyError:
            return None
    
    boarding_new["오름혼잡도"] = boarding_new.apply(lambda row: get_congestion_value(row, "오름차순"), axis=1)
    boarding_new["내림혼잡도"] = boarding_new.apply(lambda row: get_congestion_value(row, "내림차순"), axis=1)

    # 전역의 혼잡도
    boarding_new["오름전역혼잡도"] = None
    boarding_new["내림전역혼잡도"] = None

    def get_previous_station_congestion_value(row, direction):
        previous_station_order = get_previous_station_order(row["호선"], row["순서"], direction)
        if previous_station_order is None:
            return None
        hosun = row["호선"]
        previous_station_number = get_station_number(hosun, previous_station_order)
        key = (row["요일구분"], row["시간대"], previous_station_number, direction)
        try:
            return congest_lookup.loc[key, "혼잡도"]
        except KeyError:
            return None

    boarding_new["오름전역혼잡도"] = boarding_new.apply(lambda row: get_previous_station_congestion_value(row, "오름차순"), axis=1)
    boarding_new["내림전역혼잡도"] = boarding_new.apply(lambda row: get_previous_station_congestion_value(row, "내림차순"), axis=1)

    # 전 역이 없는 경우 Drop
    boarding_new = boarding_new.dropna()
    
    return boarding_new
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import Ridge, Lasso
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import numpy as np

# 1. Feature/Target 지정
# 오름전역혼잡도, 승차, 하차, 시간대 -> X
# 오름혼잡도 -> y
features = ['오름전역혼잡도', '승차', '하차', '시간대']
target = '오름혼잡도'

X = merged_non_transfer[features].copy()
y = merged_non_transfer[target].copy()

# 2. 결측치 처리
X = X.fillna(X.median(numeric_only=True))
y = y.fillna(y.median())

# 3. 이상치(아웃라이어) 클리핑 (IQR 방식)
Q1 = X.quantile(0.25)
Q3 = X.quantile(0.75)
IQR = Q3 - Q1
X = np.clip(X, Q1-1.5*IQR, Q3+1.5*IQR)

# 4. 정규화/표준화: 수치형 데이터라 표준화가 일반적
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 5. 데이터 분리
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# 6. 모델 정의 및 하이퍼파라미터 튜닝 (예시: RandomForest, Ridge)
models = {
    'RandomForest': RandomForestRegressor(random_state=42),
    'Ridge': Ridge(random_state=42),
    'Lasso': Lasso(random_state=42)
}
params = {
    'RandomForest': {'n_estimators': [100, 200], 'max_depth': [None, 8, 16]},
    'Ridge': {'alpha': [0.01, 0.1, 1, 10]},
    'Lasso': {'alpha': [0.01, 0.1, 1, 10]}
}

search_results = {}
for name, model in models.items():
    grid = GridSearchCV(model, params[name], scoring='neg_mean_squared_error', cv=5, n_jobs=-1)
    grid.fit(X_train, y_train)
    best_model = grid.best_estimator_
    y_pred = best_model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    search_results[name] = {
        'Best Params': grid.best_params_,
        'MSE': mse,
        'MAE': mae,
        'R2': r2,
        'Model': best_model
    }
    print(f"{name}: Best Params={grid.best_params_} | MSE={mse:.3f} | MAE={mae:.3f} | R2={r2:.3f}")

# 성능 비교 결과 DataFrame화
import pandas as pd
pd.set_option('display.precision', 3)
performance_df = pd.DataFrame([
    {'Model': k, **{kk: v for kk, v in val.items() if kk != 'Model'}}
    for k, val in search_results.items()
])
print("\n모델별 성능 비교:")
display(performance_df)

# 실전 적용시 테스트셋 예측 결과/실측 plot, 잔차 분석, feature importance(가능한 경우) 등도 반드시 확인
import matplotlib.pyplot as plt

best_model_name = performance_df.sort_values("R2", ascending=False).iloc[0]["Model"]
final_model = search_results[best_model_name]["Model"]
y_final_pred = final_model.predict(X_test)

plt.figure(figsize=(7, 7))
plt.scatter(y_test, y_final_pred, alpha=0.6)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.xlabel("실제 오름혼잡도")
plt.ylabel("예측 오름혼잡도")
plt.title(f"실제 vs 예측 오름혼잡도 ({best_model_name})")
plt.show()

# feature importance (RandomForest인 경우만)
if best_model_name == 'RandomForest':
    importances = final_model.feature_importances_
    indices = np.argsort(importances)[::-1]
    plt.bar([features[i] for i in indices], importances[indices])
    plt.title("Feature Importances (RandomForest)")
    plt.show()
```
